## 1. Separate 'Hero' and 'Publisher' into two columns.

## 2. Unpack the 'Measurements' column of dictionaries into separate columns

- We will to convert the single "Measurements" column into 2 separate columns "Height" and "Weight".

- First we will replace the single quotes in string-dictionaries with double quotes so we can use json.loads

```python
import json
## use .str.replace to replace all single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
height_weight = df['Measurements'].apply(pd.Series)
height_weight
```

## 3. Change the titles names of 'Height' and 'Weight' columns to 'Height(cm)' and 'Weight(kg)' and then remove 'cm' and 'kg' from the values in each column so we can have only numbers.

```python
## save the 2 new columns into the dataframe with new title names
df[['Height (cm)','Weight (kg)']] = df['Measurements'].apply(pd.Series)
df.head()
```

```python
# Make a list of all characters to replace
to_replace = ['cm','kg']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    df['Height (cm)'] = df['Height (cm)'].str.replace(char,'',regex=False)
    df['Weight (kg)'] = df['Weight (kg)'].str.replace(char,'',regex=False)
df.head()
```

Let's drop the original 'Measurements' column

```python
df = df.drop(columns=['Measurements'])
df.head()
```

```python
# Check to see if the 'Height' and 'Weight' columns are numeric
df.info()
```

```python
# Change the 'Height' and 'Weight' columns to float64 datatype
df['Height (cm)'] = df['Height (cm)'].astype('float64')
df['Weight (kg)'] = df['Weight (kg)'].astype('float64')
df.info()
```

- The 'Height' and 'Weight' columns are now floats which is a numeric datatype.

## 5. Convert the 'Powers' column into one-hot-encoded columns (a column for each item in the list)

```python
df1['Powers_split'].value_counts()
```

- We first need to get a list of all of the unique options that appear in this column.


- We will then loop through these unique options and leverage .str.contains to create a new column for each. 

```python
## exploding the column of lists
exploded = df1.explode('Powers_split')
exploded[['hero_names','Powers', 'Powers_split']].head()
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

We will now use a For Loop and .str.contains to create the new columns

```python
for col in cols_to_make:
    df1[col] = df1['Powers'].str.contains(col)
df1
```

- Now that we have one-hot-encoded columns for every power that appears in the dataset, we can delete the following columns: 'hero_names', 'Powers' and 'Powers_split'

```python
df1 = df1.drop(columns=['Powers', 'Powers_split'])
df1.head()
```

```python
## merging with different column names
merged = pd.merge(df, df1, left_on='Hero', right_on='hero_names')
merged
```

## 6. Use your combined DataFrame to answer the following questions.

a. Compare the average weight of super powers who have Super Speed to those who do not.

b. What is the average height of heroes for each publisher?

### a. Compare the average weight of super powers who have Super Speed to those who do not.

```python
Power_weight = merged.groupby('Super Speed')['Weight (kg)'].mean()
Power_weight
```

- The 'True' value above represents the average weight of super powers who have Super Speed. The 'False' value represents the average weight of super powers who do not have Super Speed.

- We can see that the average weight (129.40 kg) of super powers who have Super Speed is greater/higher than the average weight (101.77 kg) of super powers who do not have Super Speed.

### b. What is the average height of heroes for each publisher?

```python
hero_height = merged.groupby('Publisher')['Height (cm)'].mean()
hero_height
```

- The above has the average height (in cm) of heroes for each publisher. The Publisher is listed on the left side and the average heights on the right side.
